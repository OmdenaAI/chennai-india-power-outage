```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
import seaborn as sb
```

```python
d1=pd.read_csv('electricity_generation.csv')
d1.head()
```

```python
d10=d1[d1['Year']==2010]
d10=d10.rename(columns={'Net Generation (M.U)':'Net Generation of 2010(M.U)'})
d10=d10.drop(['Year'],axis=1)
d11=d1[d1['Year']==2011]
d11=d11.rename(columns={'Net Generation (M.U)':'Net Generation of 2011(M.U)'})
d11=d11.drop(['Year'],axis=1)
d12=d1[d1['Year']==2012]
d12=d12.rename(columns={'Net Generation (M.U)':'Net Generation of 2012(M.U)'})
d12=d12.drop(['Year'],axis=1)
d13=d1[d1['Year']==2013]
d13=d13.rename(columns={'Net Generation (M.U)':'Net Generation of 2013(M.U)'})
d13=d13.drop(['Year'],axis=1)
d16=d1[d1['Year']==2016]
d16=d16.rename(columns={'Net Generation (M.U)':'Net Generation of 2016(M.U)'})
d16=d16.drop(['Year'],axis=1)
d17=d1[d1['Year']==2017]
d17=d17.rename(columns={'Net Generation (M.U)':'Net Generation of 2017(M.U)'})
d17=d17.drop(['Year'],axis=1)
d18=d1[d1['Year']==2018]
d18=d18.rename(columns={'Net Generation (M.U)':'Net Generation of 2018(M.U)'})
d18=d18.drop(['Year'],axis=1)
d19=d1[d1['Year']==2019]
d19=d19.rename(columns={'Net Generation (M.U)':'Net Generation of 2019(M.U)'})
d19=d19.drop(['Year'],axis=1)
d20=d1[d1['Year']==2020]
d20=d20.rename(columns={'Net Generation (M.U)':'Net Generation of 2020(M.U)'})
d20=d20.drop(['Year'],axis=1)
d21=d1[d1['Year']==2021]
d21=d21.rename(columns={'Net Generation (M.U)':'Net Generation of 2021(M.U)'})
d21=d21.drop(['Year'],axis=1)
d22=d1[d1['Year']==2022]
d22=d22.rename(columns={'Net Generation (M.U)':'Net Generation of 2022(M.U)'})
d22=d22.drop(['Year'],axis=1)
```

```python
d1.groupby(['Year']).size()
```

```python
d10.head()
```

```python
#d11.reset_index(inplace=True)
d11.head()
```

```python
d2=pd.merge(d10,d11,on=['Type','Name'],how='left')
d2.head()
```

```python
d2=pd.merge(d10,d11,on=['Type','Name'],how='left')
d3=pd.merge(d2,d12,on=['Type','Name'],how='left')
d4=pd.merge(d3,d13,on=['Type','Name'],how='left')
d5=pd.merge(d4,d16,on=['Type','Name'],how='left')
d6=pd.merge(d5,d17,on=['Type','Name'],how='left')
d7=pd.merge(d6,d18,on=['Type','Name'],how='left')
d8=pd.merge(d7,d19,on=['Type','Name'],how='left')
d8.head()
```

```python
print(d8['Net Generation of 2011(M.U)'].isna().sum(),d8['Net Generation of 2012(M.U)'].isna().sum(),d8['Net Generation of 2013(M.U)'].isna().sum(),d8['Net Generation of 2016(M.U)'].isna().sum(),d8['Net Generation of 2017(M.U)'].isna().sum(),d8['Net Generation of 2018(M.U)'].isna().sum(),d8['Net Generation of 2019(M.U)'].isna().sum())
```

```python
da=pd.merge(d20,d21,on=['Type','Name'],how='left')
db=pd.merge(da,d22,on=['Type','Name'],how='left')
db.head()
```

```python
print(db['Net Generation of 2021(M.U)'].isna().sum(),db['Net Generation of 2022(M.U)'].isna().sum())
```

```python
d9=pd.merge(d8,db,on=['Type','Name'],how='left')
d9.head()
```

```python
print(d9['Net Generation of 2020(M.U)'].isna().sum(),d9['Net Generation of 2021(M.U)'].isna().sum(),d9['Net Generation of 2022(M.U)'].isna().sum())
```

```python
dc=d9.iloc[:,2:]
```

```python
dc.head()
```

```python
dc.info()
```

```python
for i in dc.columns:
    dc[i]=dc[i].abs()
```

```python
dc=dc.apply(lambda row: row.fillna(row.mean()),axis=1)
#dc.fillna(dc.mean(axis=1),axis=1)
```

```python
dc.head()
```

```python
dc['Net Generation of 2014(M.U)']=dc.mean(axis=1)
dc['Net Generation of 2015(M.U)']=dc.mean(axis=1)
```

```python
dc['Type']=d9['Type']
dc['Name']=d9['Name']
```

```python
dc=dc.round(2)
```

```python
dc=dc[['Type','Name','Net Generation of 2010(M.U)', 'Net Generation of 2011(M.U)',
       'Net Generation of 2012(M.U)', 'Net Generation of 2013(M.U)',
       'Net Generation of 2014(M.U)', 'Net Generation of 2015(M.U)',
       'Net Generation of 2016(M.U)', 'Net Generation of 2017(M.U)',
       'Net Generation of 2018(M.U)', 'Net Generation of 2019(M.U)',
       'Net Generation of 2020(M.U)', 'Net Generation of 2021(M.U)',
       'Net Generation of 2022(M.U)']]
```

```python
dc.head()
```

```python
dc.to_csv('final.csv',index=False)
```

```python
df_unpivot = pd.melt(dc, id_vars=['Type','Name'], value_vars=['Net Generation of 2010(M.U)',
       'Net Generation of 2011(M.U)', 'Net Generation of 2012(M.U)',
       'Net Generation of 2013(M.U)', 'Net Generation of 2014(M.U)',
       'Net Generation of 2015(M.U)', 'Net Generation of 2016(M.U)',
       'Net Generation of 2017(M.U)', 'Net Generation of 2018(M.U)',
       'Net Generation of 2019(M.U)', 'Net Generation of 2020(M.U)',
       'Net Generation of 2021(M.U)', 'Net Generation of 2022(M.U)'])
```

```python
df_unpivot.head()
```

```python
df_unpivot['Year']=df_unpivot['variable'].str[18:22]
```

```python
df_unpivot.to_csv('final111.csv')
```

```python
mpl.plot(df_unpivot['Year'], df_unpivot['value'], label =df_unpivot['Type'])
# Function add a legend  
mpl.legend()
  
# function to show the plot
mpl.show()
```
